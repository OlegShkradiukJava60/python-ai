```python
import numpy as np
import matplotlib.pyplot as plt
from life_game_matrix import getInitialMatrix, getNextMatrix
from IPython.display import clear_output, display
SHAPE_0 = 20
SHAPE_1 = 20
LIMIT = 100
img: np = getInitialMatrix(SHAPE_0, SHAPE_1)
img_next: np = None
iteration: int = 0
plot = plt.imshow(img, cmap="gray", vmin=0, vmax=1)
while not np.array_equal(img, img_next := getNextMatrix(img)) and iteration < LIMIT:
    img = img_next
    plot.set_array(img)
    display(plt.gcf())
    clear_output(wait=True)
    iteration += 1
  
```
